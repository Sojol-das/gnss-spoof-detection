```python
# Load test data
test = pd.read_csv("../data/test.csv", low_memory=False)

# Fix channel column
test['channel'] = test['channel'].str.replace('ch','').astype(int)

# Convert numeric columns
cols = ['PRN','Carrier_Doppler_hz','Pseudorange_m','RX_time','TOW',
        'Carrier_phase','EC','LC','PC','PIP','PQP','TCD','CN0']

for c in cols:
    test[c] = pd.to_numeric(test[c], errors='coerce')

test = test.dropna()

# Create same engineered features as training
test['CN0_mean_time'] = test.groupby('time')['CN0'].transform('mean')
test['doppler_mean_time'] = test.groupby('time')['Carrier_Doppler_hz'].transform('mean')
test['pseudo_mean_time'] = test.groupby('time')['Pseudorange_m'].transform('mean')

test['corr_diff'] = test['EC'] - test['LC']

# Remove leakage features
test = test.drop(['RX_time','TOW'], axis=1)

# Make sure feature order matches training
X_test = test[X.columns]

# Predict probabilities
test_prob = model.predict_proba(X_test)[:,1]

# Apply lower threshold for imbalanced data
test_pred = (test_prob > 0.3).astype(int)

test['pred'] = test_pred

# Convert channel-level predictions to time-level
time_pred = test.groupby('time')['pred'].max().reset_index()

# Load required submission format
sample = pd.read_csv("../data/submission_format.csv")

sample['Spoofed'] = time_pred['pred']
sample['Confidence'] = 1

# Save final submission file
sample.to_csv("../submission_final.csv", index=False)

# Quick check
print(sample.head())
print(sample['Spoofed'].value_counts())
```
